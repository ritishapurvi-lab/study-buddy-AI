```python
subject=input("which subject are you studying? ")
print("Great!📚 you're studying",subject,"Let's make a plan for it.")
```

```python
chapters=int(input("How many chapters are there? "))
days=int(input("How many days until your exam is conducted? "))
```

```python
print("=========================")
print("      STUDY DETAILS")
print("=========================")

print("Subject: ",subject)
print("Chapters: ",chapters)
print("Days left: ",days)
```

```python
chap_per_day=(chapters//days)
print("You should aim for studying ",chap_per_day,"chapter(s) per day.")
```

```python
start_chapter = 1
for day in range(1, days + 1):
    end_chapter = start_chapter + chap_per_day - 1

    print(f"Days",day,"Study Chapters", start_chapter,"to", end_chapter)

    start_chapter = end_chapter + 1
```
